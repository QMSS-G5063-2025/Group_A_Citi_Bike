```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load csv
july_sampled = pd.read_csv('july_sampled.csv')
```

```python
sns.set(style="darkgrid")

plt.figure(figsize=(10, 6))

#plot the hexbin
plt.hexbin(
    july_sampled['distance_km'], 
    july_sampled['duration_min'], 
    gridsize=50, 
    cmap='plasma', 
    bins='log'
)

plt.colorbar(label='log10(Number of Rides)')
plt.xlabel('Distance (km)')
plt.ylabel('Duration (minutes)')

plt.tight_layout()
plt.show()
```
