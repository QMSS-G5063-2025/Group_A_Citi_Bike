```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load csv
monthly_counts_df = pd.read_csv('monthly_counts_df.csv')
```

```python
sns.set(style="white")

# Set up plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_counts_df, x='month_label', y='count_divide', marker='o', linewidth=2, color='black')

#background shading based on seasons
seasons = {
    'Winter': ['Jan', 'Feb', 'Dec'],
    'Spring': ['Mar', 'Apr', 'May'],
    'Summer': ['Jun', 'Jul', 'Aug'],
    'Fall':   ['Sep', 'Oct', 'Nov']
}

#got hex colors according to seasons
season_colors = {
    'Winter': '#cce5ff',
    'Spring': '#d4edda',
    'Summer': '#fff3cd',
    'Fall':   '#f8d7da'
}

#creating month order
month_order = monthly_counts_df['month_label'].tolist()

#background coloring
for season, months in seasons.items():
    start = month_order.index(months[0])
    end = month_order.index(months[-1])
    plt.axvspan(start - 0.5, end + 0.5, color=season_colors[season], alpha=0.5, label=season)


#;lot
plt.ylim(1.5, 6)
plt.legend(loc='upper left', frameon=True, title="Season", fontsize=10)

plt.xlabel("Month", fontsize=16)
plt.ylabel("Number of Rides (by millions)", fontsize=16)
plt.xticks(rotation=45, fontsize=13)
plt.yticks(fontsize=13)

#labels
for i, row in monthly_counts_df.iterrows():
    plt.text(
        x=i, 
        y=row['count_divide'] + 0.27, 
        s=f"{row['count_divide']:.2f}", 
        ha='center', 
        fontsize=13
    )

plt.tight_layout()
plt.show()

```
