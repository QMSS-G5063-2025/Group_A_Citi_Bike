```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import cm
```

```python
#load csv
july_sampled = pd.read_csv('july_sampled.csv')
```

```python
#change colors to match theme
sns.set(style="darkgrid")
coolwarm = cm.get_cmap('coolwarm')
colors = [coolwarm(0.1), coolwarm(0.9)]  # One from the cool side, one from the warm

#map to member type
custom_palette = {
    'member': colors[0],
    'casual': colors[1]
}

plt.figure(figsize=(10, 6))
sns.histplot(
    data=july_sampled, 
    x='duration_min', 
    hue='member_casual', 
    bins=60, 
    element='bars', 
    stat='density', 
    multiple='dodge', 
    common_norm=False,
    palette=custom_palette
)

plt.xlabel('Duration (minutes)')
plt.ylabel('Percentage of Total Rides (normalized to user type)')
plt.legend(title='User Type', labels=['Member', 'Casual'])
plt.xlim(0, 75)
plt.tight_layout()
plt.show()
```
